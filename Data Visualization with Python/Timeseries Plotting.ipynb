# Timeseries Plotting

- A measure of unit change over time for any variable that is under observation. Forecasting can be done using historical data. It may have trend, seasonality and cyclic nature.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
# this lets plots to be plotted in same window
%matplotlib inline
rcParams['figure.figsize'] = 10, 5 # 5 inch cross 4 inch figures
sb.set_style('whitegrid')
```

```python
df = pd.read_csv('Superstore-Sales.csv', index_col = 'Order Date', parse_dates = True, encoding='cp1252')
df.head()
```

```python
# Plotting order quantity
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n=100, random_state =25, axis =0)
plt.xlabel('Order Date')
plt.ylabel('Order Qty')
plt.title('Superstore-Sales')
df2['Order Quantity'].plot()
```
