# Line, Pie and Bar Charts

- Line Chart: to show changes over time Bar Chart: to show changes by category 


- Pie Chart: to show categorical data as proportions of whole

- There are 2 methods to build plots in Python. One is Functional(calling plot function on variable or more variables) and other is Object-Oriented where you first build a blank object and then plot a figure on it.


- Most Popular Data Visualization Libraries are Seaborn and matplotlib.

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
# this lets plots to be plotted in same window
%matplotlib inline
rcParams['figure.figsize'] = 5, 4 # 5 inch cross 4 inch figures
sb.set_style('whitegrid')
```

# Line Chart

```python
# Create sample dataset
x = range(1,10)
y = [1,2,5,8,3,6,2,0,9]
# Plot data using matplotlib
plt.plot(x,y)
```

```python
cars = pd.read_csv('mtcars.csv')
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
mpg = cars['mpg']
```

```python
# ploting using pandas
mpg.plot()
```

```python
# Plotting multiple variables
# Create subset of the dataset
df = cars[['cyl','mpg','wt']]
```

```python
df.plot()
```

# Bar Plots

```python
plt.bar(x,y)
```

```python
# using pandas for bar chart
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

# Pie Chart

```python
x= [1,2,4,3,0,5]
plt.pie(x)
plt.show()
```

```python
# Saving plots as jpeg
plt.savefig('PieChart.jpeg')
plt.show()
```
