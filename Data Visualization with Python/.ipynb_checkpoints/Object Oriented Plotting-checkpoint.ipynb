
# Object Oriented Plotting

Steps:

    Create blank object
    Add axes
    Generate plots within the figure
    Specify plotting and layout parameters for the plots within the figure



```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from numpy.random import randn

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
# this lets plots to be plotted in same window
%matplotlib inline
rcParams['figure.figsize'] = 5, 4 # 5 inch cross 4 inch figures
```

```python
# Create sample dataset
x = range(1,10)
y = [1,2,5,8,3,6,2,0,9]

# Create blank figure
fig = plt.figure()
ax = fig.add_axes([.1,.1, 1,1])
ax.plot(x,y)
```

```python
## Adding tick marks

fig = plt.figure()
ax = fig.add_axes([.1,.1, 1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,2,4,6,8,10])
ax.set_yticks([1,2,3,4,5])
ax.plot(x,y)

```

```python
## Adding Grid Lines

fig = plt.figure()
ax = fig.add_axes([.1,.1, 1,1])
ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

```python
# Generate Subplots
fig = plt.figure()
fig,(ax1,ax2) = plt.subplots(1,2)
ax1.plot(x)
ax2.plot(y)
```
